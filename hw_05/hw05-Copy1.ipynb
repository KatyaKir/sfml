```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv', sep='\t', index_col='id')
```

```python
train['text']= train['name'] +' '+ train['description']
```

```python
test = pd.read_csv('test.csv', sep='\t', index_col='id')
```

```python
test['text']= test['name'] +' '+ test['description']
```

```python
y = train['target']
```

```python
train = train.drop('target', axis=1)
```

```python
all_data = pd.concat([train, test])
```

```python
all_text = list(all_data['text'])
```

```python
from bs4 import BeautifulSoup
```

```python
texts_without_tags = []
for t in all_text[:50000]:
    texts_without_tags.append(
        BeautifulSoup(t, 'lxml').text  
    )
```

```python
import re
import pymystem3
mystem = pymystem3.Mystem()
```

```python
def my_tokenizer(text):
    lemms = mystem.lemmatize(text)
    return [x for x in lemms if re.match('^[а-яё]+$', x)]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(tokenizer=my_tokenizer)
```

```python
m = cv.fit_transform(texts_without_tags)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(m, y[:50000], test_size=0.2, random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
Train = pd.read_csv('train.csv', sep='\t', index_col='id')
```

```python
Train['text']= Train['name'] +' '+ Train['description']
```

```python
Test = pd.read_csv('test.csv', sep='\t', index_col='id')
```

```python
Test['text']= Test['name'] +' '+ Test['description']
```

```python
Y = Train['target']
```

```python
Train = Train.drop('target', axis=1)
```

```python
All_data = pd.concat([Train, Test])
```

```python
All_text = list(All_data['text'])
```

```python
texts_without_tags_full = []
for s in All_text[:]:
    texts_without_tags_full.append(
        BeautifulSoup(s, 'lxml').text  
    )
```

```python
Mystem = pymystem3.Mystem()
```

```python
def My_tokenizer(text):
    lemms = Mystem.lemmatize(text)
    return [x for x in lemms if re.match('^[а-яё]+$', x)]
```

```python
CV = CountVectorizer(tokenizer=My_tokenizer)
```

```python
nn = CV.fit_transform(texts_without_tags_full)
```

```python
#LogisticRegression
lr_test = LogisticRegression(C=1, penalty='l1', random_state=10)
lr_test.fit(nn[:200000], Y)
```

```python
aaa = lr_test.predict_proba(CV.transform(Test['text']))[:,1]
```
